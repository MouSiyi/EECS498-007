## Why Neural Network?
### Brain Analogy
dendrites() -> cell body -> axon() ->presynaptic terminal




### Space Warping
For points that are not linearly separatable in original space, after linear transform, still can not be linearly separatable.

### Universal Approximation
A neural network with one hidden layer can approximate any function: $R^N -> R^M$ with arbitrary precision




Neural nets can represent any function.
But
-  We still do not know whether we can actually learn any function with SGD

## Optimization
### Convex Functions

- Bowl shape
- Easy and robust to optimize: for convex functions, we can derive theoretically that they converge to global minimum.    

And loss func for linear classifier are convex         
        
But most neural networks need nonconvex optimization
- Few or no guarantees for convergence
- But empirically it seems to work anyway, thus requres much deeper research
